```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

house = pd.read_table('AmesHousing.txt')
house.info()
```

```python
drop_cols = ['Alley', 'Fireplace Qu', 'Pool QC', 'Fence', 'Misc Feature', 'Lot Frontage']
house = house.drop(drop_cols, axis = 1)
house = house.dropna(axis = 0)
house.isnull().sum()
```

```python
house.info()
```

```python
house = house.select_dtypes(include = ['int64', 'float64'])
house.info()
```

```python
train = house.iloc[0:1460]
test = house.iloc[1460: ]
```

## define functions 

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np

train_cols = house.columns[0:-1]
target_cols = ['SalePrice']

def transform_features(train):
    return train

def select_features():
    return train['Fr Liv Area', 'SalePrice']

def train_and_test(train, test):
    train_X = train[train_cols]
    train_Y = train[target_cols]
    test_X = test[train_cols]
    test_Y = test[target_cols]
    model = LinearRegression()
    kf = KFold()
    mses = cross_val_score(model, house[train_cols], house[target_cols], scoring='neg_mean_squared_error', cv=kf)
    rmses = np.sqrt(np.power(mses, 2))
    avg_rmse = np.mean(rmses)
    return (rmses, avg_rmse)

```
