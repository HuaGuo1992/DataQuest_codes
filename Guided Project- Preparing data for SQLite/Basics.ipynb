```python
import pandas as pd

award_data = pd.read_csv('academy_awards.csv', encoding = "ISO-8859-1")
award_data.info()
```

```python
award_data.head()
```

```python
drop_cols = ['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']
award_data = award_data.drop(drop_cols, axis = 1)
```

```python
award_data = award_data.dropna(axis = 0)
```

```python
award_data['Year'] = award_data['Year'].str[0:4].astype(int)
```

```python
later_than_2000 = award_data[award_data['Year'] > 2000]
later_than_2000.sample(10)
```

```python
category = ['Actor -- Leading Role', 'Actor -- Supporting Role', 'Actress -- Leading Role', 'Actress -- Supporting Role']
nominations = award_data[award_data['Category'].isin(category)]
nominations.info()
```

```python
nominations.sample(10)
```

```python
won_map = {'NO':0, 'YES':1}
nominations['Won'] = nominations['Won?'].map(won_map)

nominations.sample(10)
```

```python
final_nominations = nominations.drop(['Won?'], axis = 1)
```
