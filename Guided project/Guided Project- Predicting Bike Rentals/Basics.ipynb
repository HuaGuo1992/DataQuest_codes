# 2. Data Cleaning 

```python
def hour_convert(row):
    v = row
    if v in range(6, 12):
        return 1
    elif v in range(12, 18):
        return 2
    elif v in range(18, 24):
        return 3
    else:
        return 4
    
bike['time_table'] = bike['hr'].apply(hour_convert)
```

# 3. Data Spliting 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

columns = list(bike.columns)
for ls in ['cnt', 'casual', 'registered', 'dteday', 'instant']:
    columns.remove(ls)

train_X, test_X, train_y, test_y = train_test_split(bike[columns], bike['cnt'], test_size=0.2)
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_X, train_y)
predictions = lr.predict(test_X)
mar = mean_absolute_error(test_y, predictions)
mar
```

```python
from sklearn.ensemble import RandomForestRegressor
drt = RandomForestRegressor(n_estimators=50, min_samples_leaf=13)
drt.fit(train_X, train_y)
drt.fit(train_X, train_y)
predictions = drt.predict(test_X)
mar = mean_absolute_error(test_y, predictions)
mar
```
