# 2. Clean data

## 2.1 numeric variable

```python
import numpy as np
car = car.replace('?', np.nan)
to_numeric = ['normalized_losses', 'bore', 'stroke', 'horsepower', 'peak_rpm', 'price']
car[to_numeric] = car[to_numeric].astype(float)
car = car.drop(['normalized_losses'],axis = 1)
car.isnull().sum()
```

```python
car = car.dropna(axis = 0)
car.isnull().sum()
```

# 3. Data training 

## 3.1 Univariate model 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

def knn_train_test(train_cols, target_cols, data):
    X = data[train_cols]
    Y = data[target_cols]
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)
    neigh = KNeighborsRegressor()
    neigh.fit(X_train, Y_train)
    predict = neigh.predict(X_test)
    rmse = (mean_squared_error(Y_test, predict))**(0.5)
    return rmse

rmse_dict = {}
for col in numeric_cols:
    rmse_dict[col] = knn_train_test([col], ['price'], car)
rmse_result_series = pd.Series(rmse_dict)
rmse_result_series.sort_values()    
```

```python
def knn_train_test(train_cols, target_cols, data):
    X = data[train_cols]
    Y = data[target_cols]
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)
    k_list = [1, 3, 5, 7, 9]
    rmse = []
    for k in k_list:
        neigh = KNeighborsRegressor(n_neighbors = k)
        neigh.fit(X_train, Y_train)
        predict = neigh.predict(X_test)
        item = (mean_squared_error(Y_test, predict))**(0.5)
        rmse.append(item)
    return rmse

rmse_dict = {}
for col in numeric_cols:
    rmse_dict[col] = knn_train_test([col], ['price'], car)
rmse_dict
```

```python
import matplotlib.pyplot as plt
for key, item in rmse_dict.items():
    plt.plot(item)
plt.show()
```

## 3.2 Mutivariate model 

```python
def knn_train_test(train_cols, target_cols, data):
    X = data[train_cols]
    Y = data[target_cols]
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)
    neigh = KNeighborsRegressor()
    neigh.fit(X_train, Y_train)
    predict = neigh.predict(X_test)
    rmse = (mean_squared_error(Y_test, predict))**(0.5)
    return rmse

cols = list(set(numeric_cols) - set(['price']))
col2 = ['engine_size', 'highway_mpg']
col3 = ['engine_size', 'highway_mpg', 'horsepower']
col4 = ['engine_size', 'highway_mpg', 'horsepower', 'curb_weight']
col5 = ['engine_size', 'highway_mpg', 'horsepower', 'curb_weight', 'city_mpg']

keys = ['all', 2, 3, 4, 5]
items = [cols, col2, col3, col4, col5]
rmse_result = {}
for key, item in zip(keys, items):
    rmse_result[key] = knn_train_test(item, ['price'], car)
rmse_result
```

```python
def knn_train_test(train_cols, target_cols, data):
    X = data[train_cols]
    Y = data[target_cols]
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)
    k_list = range(1, 26)
    rmse = []
    for k in k_list:
        neigh = KNeighborsRegressor(n_neighbors = k)
        neigh.fit(X_train, Y_train)
        predict = neigh.predict(X_test)
        item = (mean_squared_error(Y_test, predict))**(0.5)
        rmse.append(item)
    return rmse

keys = ['all', 'two', 'three', 'four', 'five']
items = [cols, col2, col3, col4, col5]
rmse_result = {}
for key, item in zip(keys, items):
    rmse_result[key] = knn_train_test(item, ['price'], car)
rmse_result
```

## 3.3 K-fold cross validation model 

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsRegressor

def knn_train_test2(train_cols, target_cols, data):
    kf = KFold(5, shuffle=True, random_state = 1)
    model = KNeighborsRegressor()
    mses = cross_val_score(model, car[train_cols], car[target_cols], scoring = 'neg_mean_squared_error', cv=kf)
    rmses = np.sqrt(np.power(mses, 2))
    return np.mean(rmses)
keys = ['all', 'two', 'three', 'four', 'five']
items = [cols, col2, col3, col4, col5]
rmse_result = {}
for key, item in zip(keys, items):
    rmse_result[key] = knn_train_test(item, ['price'], car)
rmse_result
```

```python
for key, item in rmse_result.items():
    plt.plot(item)
plt.show()
```
